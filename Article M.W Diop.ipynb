```python
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_excel('C:/Users/pc/Documents/Article Mor Wéllé/Base article RSE.xlsx')
df
```

# Carte des corrélations améliorée

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(16 , 16))
mask = np.triu(np.ones_like(df.corr(), dtype = bool))
sns.heatmap(df.corr(), mask = mask, center = 0,
           cmap ='RdBu', linewidths = 1,
           annot =True, fmt = '.2f' , vmin = -1, vmax = 1)
plt.title('Carte des corrélations de Pearson')
plt.show()
```
